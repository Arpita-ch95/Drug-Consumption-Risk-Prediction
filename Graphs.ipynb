```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#Display related options
%matplotlib inline
# palette options: "colorblind", 'Set1', "Set2", "Set3", "summer", "spring", "pastel", "rainbow"

sns.set(palette = "rainbow")

pd.set_option('display.precision', 2)
```

```python
#Read dataframe in pandas
behaviour = pd.read_csv("Behavioural Score.csv")
behaviour.head(10)
```

```python
behavioural_corr = behaviour.corr()
behavioural_corr
```

```python
# cmap is very much like color palette
#sns.heatmap(data=nba_corr , annot = True , cmap ="Set1" )

#annot = annotation to give column names on X and Y axis
#cmap = colour map
sns.heatmap(data=behavioural_corr, annot = True, cmap = "coolwarm");
```

```python
#Pearson Correlation Interpretation
#Values near or = +1: Positively corelated
#Values near 0: No correlation
#Values near or = -1: Negatively correlated


#Correlation between behavioural data between:

#SS and Impulsive = 0.62
#Cscore and Escore = 0.31
#SS and Oscore = 0.42
```

```python
demographic = pd.read_csv("Demo.csv")
demographic.head(10)
```

```python
demographic["Age"] = pd.Categorical(demographic["Age"])
demographic["Age Code"] = demographic["Age"].cat.codes
demographic["Age Code"].corr(demographic["Nicotine Binary"])
```

```python
demographic["Country"] = pd.Categorical(demographic["Country"])
demographic["Country Code"] = demographic["Country"].cat.codes
demographic["Country Code"].corr(demographic["Cannabis Binary"])
```

```python
demographic["Ethnicity"] = pd.Categorical(demographic["Ethnicity"])
demographic["Ethnicity Code"] = demographic["Ethnicity"].cat.codes
demographic["Ethnicity Code"].corr(demographic["Cannabis Binary"])
```
